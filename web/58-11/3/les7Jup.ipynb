работа с сеаборном

```python
import seaborn as sns
stores = pd.read_csv('stores.csv')
sns.lmplot(x='Store', y='Size', data=stores)
plt.show()
```

fit_reg - включение или отключение регрессионной прямой.

```python
gr = sns.lmplot(x='Store', y='Size', hue='Type', data=stores, fit_reg=True)
gr.set_xlabels('Магазин')
gr.set_ylabels('Размер')
plt.show()
```

задание 1.

```python
data = train.query('stores == 1')
sns.lmplot(x='Dept', y='Weekly_Sales', hue='IsHoliday', data=data)
plt.show()
```

задание 3.


```python
data = train.query('stores == 1').groupby(
    'Dept', as_index=False).aggregate({'Weekly_Sales':'mean'})
sns.scatterplot(x='Dept', y='Weekly_Sales', data=data)
plt.show()
```

```python
mean = train.groupby(['stores', 'IsHoliday'],
                     as_index=False).aggregate({'Weekly_Sales': 'mean'})
min_ = train.groupby(['stores', 'IsHoliday'],
                     as_index=False).aggregate({'Weekly_Sales': 'min'})
max_ = train.groupby(['stores', 'IsHoliday'],
                     as_index=False).aggregate({'Weekly_Sales': 'max'})

sns.scatterplot(x='stores', y='Weekly_Sales', hue='IsHoliday', data=max_)
sns.scatterplot(x='stores', y='Weekly_Sales', hue='IsHoliday', data=min_)
sns.scatterplot(x='stores', y='Weekly_Sales', hue='IsHoliday', data=mean)
```
