делаем генерацию из 1000 символов в сидом 100

```python

import numpy
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats  # модуль статистических функций
import scipy as sp
import statistics
import random
random.seed(100)
n = 1000
d = [random.randint(0, 30) for i in range(n)]
print(d)
```

```python
c = [0,1/4,1/2,3/4,1]

print(numpy.quantile(d, q=c))
```

```python
c = [0.25, 0.5, 0.75]

print(numpy.quantile(d, q=c))
```
