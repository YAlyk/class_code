```python
import matplotlib.pyplot as plt

d_x = [1, 2, 3, 4, 5, 2]
d_y = [1, 4, 9, 16, 25, 10]

plt.scatter(d_x, d_y)
plt.show()
```

```python
def y(x):
    return x **2

x_min = -10
x_max = 10

data_x = [x for x in range(x_min,x_max+1,1)]
data_y = list(map(y,data_x))

plt.scatter(data_x,data_y)
plt.show()
```
